```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load the dataset from CSV
data = pd.read_csv('/Users/muhammadthoriqzam/Downloads/teachingMLDL-main/01. Machine Learning/02. Week 2/Dataset/FinancialMarket.csv')
```

```python
# Extract the x and combined data columns
x = data['x'].values
combined_data = data['combined_data'].values

# Step 3: Visualize the dataset
plt.figure(figsize=(7, 6))

# Plot the combined data
plt.plot(x, combined_data, label='company’s stock price', color='blue')

# Add labels and title
plt.title('Financial Market Response to Earnings Reports')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()  # Ensure there's a label to show in the legend

# Display the plot
plt.show()
```
